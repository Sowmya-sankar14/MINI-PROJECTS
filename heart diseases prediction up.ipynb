```python
import sklearn
from sklearn.model_selection import train_test_split
x_training_set,x_testing_set,y_training_set,y_testing_set=train_test_split(x,y,test_size=0.25,random_state=83)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dec=DecisionTreeClassifier()
```

```python
dec.fit(x_training_set,y_training_set)
```

```python
y_pred=dec.predict(x_testing_set)
y_pred
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_testing_set,y_pred )
score
```
